```python
import pandas as pd
from fbprophet import Prophet
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_data4.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
df_data4
```

```python
df_data4.columns=["ds","y"]
```

```python
df_data4.plot()
```

```python
df_data4.head()
```

```python
df_data4['y'].plot()
```

```python
df_data4['y']=df_data4['y']-df_data4['y'].shift(1)
```

```python
df_data4.head()
```

```python
df_data4['y'].plot()
```

```python
model=Prophet()
model.fit(df_data4)
```

```python
model
```

```python
model.seasonalities
```

```python
model.component_modes
```

```python
future_dates=model.make_future_dataframe(periods=170)
```

```python
future_dates
```

```python
prediction=model.predict(future_dates)
```

```python
prediction.head()
```

```python
prediction[['ds','yhat','yhat_lower','yhat_upper']].tail(170)
```

```python
model.plot(prediction)
```
